```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# seu DataFrame já existente
df = pd.read_csv('../Magnetizacao_media.txt', header=None, names=['Mag'])

# lê o segundo arquivo
energia = pd.read_csv('../Energia.txt', header=None)

# adiciona como nova coluna
df['Energia'] = energia[0]

print(df)
```

```python
df.head()
```

```python
plt.figure(figsize=(18,6))
plt.scatter(df["Energia"], df["Mag"])

plt.xlabel("Energia")
plt.ylabel("Magnetização")
```

```python
plt.figure(figsize=(18,6))
plt.plot(df["Energia"], df["Mag"])

plt.xlabel("Energia")
plt.ylabel("Magnetização")
```

```python
plt.figure(figsize=(18,6))
plt.plot(df.index, df["Mag"])

plt.xlabel("x")
plt.ylabel("Magnetização")
```

```python
plt.figure(figsize=(18,6))
plt.plot(df.index, df["Energia"])

plt.xlabel("x")
plt.ylabel("Energia")
```
